- 자기자신을 다시 호출하는 함수 호출 방법
    - 장점 : 코드를 아주 짧게 작성. 수학적 방법이라 쉽게 이해
    - 단점 : 디버깅 어렵다. 수학 못하면 이해가 어렵다. 속도저하!

```python
count = 0

def openBox():
    global count
    print(f'상자를 엽니다. {count}')
    count += 1
    openBox() # 재귀호출

openBox()
```

- 위의 코드 실행시 count가 2969정도 되면 아래 예외 발생
    - RecursionError: maximum recursion depth exceeded while calling a Python object
    - 파이썬에서는 무한 재기호출을 방지

```python
count = 5

def openBox2():
    global count
    print(f'상자를 엽니다. {count}')
    count -= 1
    if count == 0:
        print('반지를 넣고 반환!')
        return
    openBox2()  # 재귀호출
    print('상자를 닫습니다.')
    # return

openBox2()
```

#### 팩토리얼
- S(n) 과 차이가 연산자가 +, *

- 반복문으로 팩토리얼 구현

```python
num = 1000
value = 1
for n in range(num, 0, -1):
    value *= n

print(f'{num}! = {value}')
```

- 재귀호출

```python
def factorial(n):
    if n <= 1:
        return 1
    return n * factorial(n-1)

print(f'{num}! = {factorial(num)}')
```
